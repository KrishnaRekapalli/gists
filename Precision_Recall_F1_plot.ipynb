```
import numpy as np 
import matplotlib.pyplot as plt
```

```
from mpl_toolkits import mplot3d
```

```
P = [0.1*i for i in range(1,11)]
R = [0.1*i for i in range(1,11)]
```

```
F1 = np.zeros((10,10))
```

```
for i,prec in enumerate(P):
  for j, recall in enumerate(R):
    F1[i][j] = 2* (prec*recall)/(prec+recall)
```

```
plt.imshow(F1.T, cmap='hot')
plt.show()
```

```
def f1(x, y):
    return 2*((x*y)/(x+y))

def am(x,y):
  return (x+y)/2

def gm(x,y):
  return np.sqrt(x,y)

x = np.linspace(0.1, 1, 10)
y = np.linspace(0.1, 1, 10)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
Z_am = am(x,y)
Z_gm = gm(x,y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z.T, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
